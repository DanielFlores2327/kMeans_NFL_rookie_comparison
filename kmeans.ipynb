First we are going to create a variablity score:

```python
passing_clean['Variabiity'] = 10 * ((14 *((1.25*passing_clean['Int'] + passing_clean['TD'])/passing_clean['Att'])) - (passing_clean['Cmp']/passing_clean['Att']))
passing_clean
```
