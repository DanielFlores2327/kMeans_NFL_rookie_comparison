First we are going to create a variablity score:

Variability Score = ((TD+INT)/Att) - (Cmp/Att)

So, the goal here is to find jameis winston basically by reward touchdowns and interceptions and deterring completions

```python
passing_clean['Variability'] = 10 * ((14 *((1.25*passing_clean['Int'] + passing_clean['TD'])/passing_clean['Att'])) - (passing_clean['Cmp']/passing_clean['Att']))
passing_clean
```

Although there are negative values, this is a rough measure of variability with more consistent, safer players towards the lesser values.
Now, our next value is just going to be passer rating which has the goal of roughly measuring productivity although this isn't perfect as college success doesn't transfer to pro success this is just for the point of creating pro comps

Part 2
Now, to start our Kmeans clustering model

Now, first let's plot our data to try to find clusters

```python
passing_clean.plot.scatter(x="Rate", y="Variability")
```

I can personally see four clusters: those with high rating low variability think of aaron rodgers, them those of medium variability high rating maybe a jared goff or cj stroud those with medium rating medium variability so like an anthony richardson and those with high rating high variability like josh allen

So, the clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, n_init="auto")
```

```python
kmeans.fit(passing_clean[['Rate', 'Variability']])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

Ok, so we have four clusters:

Cluster 0 is made up of quarterbacks with average passer rating and low variability, think of just a normal starter-level quarterback

Cluster 1 is made up of quarterbacks with above-average passer rating and low variability, so these are your good-solid game managers, like an alex smith

Cluster 2 is made up of basically backup-level not great players, with low rating and high variability

Cluster 3 is made up of complete studs that may also be risk-takers with high passer rating and high variability
